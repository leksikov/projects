```python
import hurst
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
import rwi
```

```python
lookback = 50
ts = hurst.random_walk(10000, proba=0.70,  min_lookback=lookback, max_lookback=lookback,)
```

```python
ts = (100+pd.Series(ts)).dropna()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(ts, columns=['price'])
```

```python
df = rwi.generate_features(df).dropna()
```

```python
df['shift_returns'] = df['returns'].shift(-1)
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.linear_model import Lasso, LinearRegression, Ridge
```

```python
df.price.plot()
```

```python
X = df.drop(columns=['shift_returns', 'price']) #  'returns', 'ROC_2', 'ROC_3'
y = df['shift_returns']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
model = Lasso(alpha=0.01) #Lasso(alpha=0.01)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
def rollingRegression(X_, y_):
    X_train_, X_test_, y_train_, y_test_ = train_test_split( X_, y_, test_size=0.20, random_state=42)
    
    model = Lasso(alpha=0.1) #Lasso(alpha=0.01) #Lasso(alpha=0.01)
    model.fit(X_train_, y_train_)
    y_pred_ = model.predict(X_test_)
    return np.sqrt(mean_squared_error(y_pred_, y_test_))
```

```python
df.shape
```

```python
p = 100
chunk = 100
res = []
for i in range(p, len(df)-p-1): # 
    
    X_rol= X.iloc[p:p+chunk]
    y_rol = y.iloc[p:p+chunk]
    try:
        res.append(rollingRegression(X_rol, y_rol))
    except:
        break
    p+=100
    print(i, p)
```

```python
df = pd.DataFrame(ts, columns=['price'])
```

```python
df = df.reset_index(drop=True)
df = df.reset_index()
df['index'] = df['index'].apply(lambda x : x // 10)
df = df.groupby('index').last()
```

```python
df= rwi.generate_features(df)
```

```python
df['shift_returns'] = df['returns'].shift(-1)
```

```python
df.dropna(inplace=True)
#df = df.fillna(0)
```

```python
X = df.drop(columns=['shift_returns', 'price']) #  'returns', 'ROC_2', 'ROC_3'
y = df['shift_returns']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
model = Lasso(alpha=0.01) #Lasso(alpha=0.01)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
plt.scatter(y_pred,y_test)
print(r2_score(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```
